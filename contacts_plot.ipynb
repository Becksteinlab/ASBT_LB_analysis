```python
# Input: Pickle files from contacts.ipynb (relinters.pkl, contacts.pkl)
# Output: PDF plot of contact distance histograms (contact_hists.pdf)
import pickle
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('relinters_of.pkl', 'rb') as handle:
    of_lip, of_prot, of_rel = pickle.load(handle)
with open('relinters_if.pkl', 'rb') as handle:
    if_lip, if_prot, if_rel = pickle.load(handle)
```

```python
with open('contacts_lip_of.pkl', 'rb') as handle:
    of_lip_counts, of_lip_bins = pickle.load(handle)
with open('contacts_prot_of.pkl', 'rb') as handle:
    of_prot_counts, of_prot_bins = pickle.load(handle)
with open('contacts_lip_if.pkl', 'rb') as handle:
    if_lip_counts, if_lip_bins = pickle.load(handle)
with open('contacts_prot_if.pkl', 'rb') as handle:
    if_prot_counts, if_prot_bins = pickle.load(handle)
```

```python
import matplotlib.gridspec as gridspec

def start_vplot(row, col, wspace=0.5, hspace=0.5, figsize=(16, 12), height_ratios='', width_ratios='', org='1d'):
    axs = []
    fig = plt.figure(figsize=figsize)
    if not height_ratios:
        height_ratios = [1] * row
    if not width_ratios:
        width_ratios = [1] * col
    gs = gridspec.GridSpec(row, col, height_ratios=height_ratios, width_ratios=width_ratios)
    gs.update(wspace=wspace, hspace=hspace)
    if org == '1d':
        for r in range(0, row):
            for c in range(0, col):
                axs.append(fig.add_subplot(gs[r, c]))
    if org == '2d':
        for r in range(0, row):
            cs = []
            for c in range(0, col):
                cs.append(fig.add_subplot(gs[r, c]))
            axs.append(cs)
    return fig, axs
```

```python
fig, axs = start_vplot(2, 2, figsize=(6, 4), org='2d', wspace=0.35, hspace=0.7)

axs[0][0].stairs(of_lip_counts, of_lip_bins, fill=True)
axs[0][0].axvline(6.0, color='black', linestyle='dashed')

axs[1][0].stairs(if_lip_counts, if_lip_bins, fill=True)
axs[1][0].axvline(6.0, color='black', linestyle='dashed')

axs[0][1].stairs(of_prot_counts, of_prot_bins, fill=True)
axs[0][1].axvline(4.5, color='black', linestyle='dashed')

axs[1][1].stairs(if_prot_counts, if_prot_bins, fill=True)
axs[1][1].axvline(4.5, color='black', linestyle='dashed')

for ax in axs:
    for a in ax:
        a.set_ylabel('g(r)')
        a.set_xlabel('r (Å)')

sns.despine(offset=0)
plt.tight_layout()
plt.savefig('contact_hists.pdf')
plt.show()
plt.close()
```
