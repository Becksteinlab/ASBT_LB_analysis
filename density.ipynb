```python
# Input: MD trajectory (structure.gro + traj.xtc)
# Output: .dx density maps for protein, ligand, lipids, and Na+ ions (in A**-3) and water (in percentage of bulk density)
import MDAnalysis as mda
from MDAnalysis.analysis.density import DensityAnalysis
import numpy as np
from MDAnalysis.analysis import align
```

```python
u_align = mda.Universe('structure.gro', 'traj.xtc')

prot = u_align.select_atoms('protein and not name H*')
dca = u_align.select_atoms('resname DCA and not name H*')
h20 = u_align.select_atoms('resname TIP3 and not name H*')
lipids = u_align.select_atoms('resname POPE and not name H* or resname POPG and not name H*')
na = u_align.select_atoms('resname SOD and not name H*')

d_prot = DensityAnalysis(prot)
d_prot.run()
d_prot.density.export("prot_density.dx", type="double")

d_dca = DensityAnalysis(dca)
d_dca.run()
d_dca.density.export("dca_density.dx", type="double")

d_h20 = DensityAnalysis(h20)
d_h20.run()
d_h20.density.convert_density('TIP4P')
d_h20.density.export("water_density.dx", type="double")

d_lipids = DensityAnalysis(lipids)
d_lipids.run()
d_lipids.density.export("lipids_density.dx", type="double")

d_na = DensityAnalysis(na)
d_na.run()
d_na.density.export("na_density.dx", type="double")
```
